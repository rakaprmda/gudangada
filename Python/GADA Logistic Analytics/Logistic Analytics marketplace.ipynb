```python
import pandas as pd
from geopy import distance
from geopy import Point
```

```python
df_seller = pd.read_csv(r"C:\Users\GADA\Documents\Python\GADA Logistic Analytics\gada_store.csv")
df_buyer = pd.read_csv(r"C:\Users\GADA\Documents\Python\GADA Logistic Analytics\hanging_fruit_marketplace_2.csv")
df_seller['point'] = df_seller.apply(lambda row: Point(latitude=row['store_latitude'], longitude=row['store_longitude']), axis=1)
df_buyer['point'] = df_buyer.apply(lambda row: Point(latitude=row['buyer_lat'], longitude=row['buyer_long']), axis=1)
df_buyer.head()
```

```python
df_buyer.head()       
```

```python
df_buyer_point = df_buyer[['point', 'buyer_code']]
df_seller_point = df_seller[['point', 'store_code']]
df_join_point = pd.DataFrame(columns = ['seller_code','buyer_code','seller_point','buyer_point','distance'])
df_buyer_point.head()
```

```python
# for index, buyer in df_buyer_point.iterrows():
#     print('1')
#     print(buyer.point)
    
for index, buyer in df_buyer_point.iterrows():
    for index, seller in df_seller_point.iterrows():
        dist = distance.distance(buyer.point,seller.point).km
        if dist < 15 :
            new_row = {'buyer_point':buyer.point, 
                       'seller_point': seller.point, 
                       'distance': dist,
                       'seller_code' : seller.store_code,
                       'buyer_code' : buyer.buyer_code }
            df_join_point = df_join_point.append(new_row, ignore_index=True)
 
```

```python
df_result = pd.merge(left=df_join_point, right=df_seller, how='left', left_on='seller_code', right_on='store_code')
df_result = pd.merge(left=df_join_point, right=df_buyer, how='left', left_on='buyer_code', right_on='buyer_code')
del df_result['seller_point']
del df_result['buyer_point']
del df_result['point']
del df_result['buyer_lat']
del df_result['buyer_long']
df_result.info()
```
